```python
import pandas as pd
```

Usaremos los métodos loc y iloc

loc[nombre_fila, nombre_columna] se basa en los nombres o etiquetas de filas y columnas para seleccionar información


loc['cero':'dos'] incluira todos los renglones entre el "cero" y el "dos" incluyendo estos mismos.


iloc[numero_fila,numero_columna] se basa en la posición numerica de filas y columnas para seleccionar información


loc[0:2] incluira todos los renglones entre el 0 y el 2 incluyendo solo el 0.

# 3 Seleccionar un rango de datos usando slice o rebanadas


Se usa la notacion stard:stop:step donde se incluyen ambos extremos

```python
# Por columnas
players = ['L. Messi', 'Cristiano Ronaldo']
df.loc[players, 'age':'club':]  # la rebanada (slice) es start = age, stop = club, step =1, va de 1 en 1 
```

```python
# por renglones
cols = ['age','dob','height_cm','weight_kg']
df.index[:10]  # no da los 10 primeros elementos
```

```python
df.loc['L. Messi':'M. Salah':,cols]  # # la rebanada (slice) es start = Messi, stop = Salah, step =1, va de 1 en 1 
```

Con iloc

```python
players = [0,1]
df.iloc[players, 1:5:1] 
```

```python
icols = range(1,5)
df.iloc[0:10:1,icols]
```

# 8 Filtrar con el método query

# 9 Método apply()

```python
df['age']
```

```python
df['age'].apply(np.sqrt)  # le vamos a sacar la raiz cuadrada a todos los valores de la columan age
```

```python
# Creamos una funcion
# una que calulara el imc de cada jugador
def calcular_imc(row):
  return row['weight_kg']/((row['height_cm']/100)**2)
df.apply(calcular_imc, axis =1)   # axis = 1 porque queremos que se ejecute a lo largo de la columna
```

# 10 Funcion Lambda

```python
# Una funcion lambda basica, que suma dos valores estaria dada por
sum_values_lambda = lambda a,b : a+b
sum_values_lambda(2,3)
```

Apply + Lambda

```python
#Usamos lambda para convertir la altura en metros
df['height_cm'].apply(lambda x:x/100)  # x es una fila de la columna (serie) height_cm
```

```python
# convertor long_name en mayus
df['long_name'].apply(lambda x: x.upper())
```

```python
# Obetner el año de nacimiento
df['dob'].apply(lambda x : x.year)
```

```python
# Obtener el imc
df.apply(lambda x : x['weight_kg']/((x['height_cm']/100)**2), axis=1)  #aqui x representa al df
```
