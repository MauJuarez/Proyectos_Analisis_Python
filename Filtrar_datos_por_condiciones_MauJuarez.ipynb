```python
import pandas as pd
df_laptops = pd.read_csv('laptop_price.csv',encoding='latin-1')
df_laptops.head(3)
```

## 1 Filtrar basado en una condición

```python
# Encontrar filas que tengan Apple en la columna company
df_laptops[df_laptops['Company'] == 'Apple']
```

```python
#verificar cuantas veces aparece la compañia Apple en la columna Company
df_laptops[df_laptops['Company'] == 'Apple'].value_counts('Company')
```

```python
# asi obtenemos cuantas veces aparece cada compañia
df_laptops.value_counts('Company')
```

```python
# Que filas no tienen HP en Company
df_laptops[df_laptops['Company'] != 'HP']
```

```python
# Encontrar las laptop con precio mayor a 2000
df_laptops[df_laptops['Price_euros'] > 2000]
```

## 2 Crear columnas basados en una condición

```python
import numpy as np
```

Haremos uso del método de numpy llamado where

np.where(condicion, si se cumple, no se cumple)

```python
np.where(df_laptops['Price_euros']>2000,'Caro','Barato')
```

```python
#Adicionamos la nueva columna que tenga esa clasificación
df_laptops['Price_category']=np.where(df_laptops['Price_euros']>2000,'Caro','Barato')
df_laptops.head(3)
```

```python
df_laptops.value_counts('Price_category')
```

```python
df_laptops['Screen_size']=np.where(df_laptops['Inches']>15,'Grande','Pequeña')
df_laptops.head(3)
```

```python
df_laptops.value_counts('Screen_size')
```

## 3 Filtrar por varias condiciones

Encontraremos las que cumplen con ser Apple y cuyo costo sea mayor que 2000

```python
df_laptops[(df_laptops['Price_euros']>2000) & (df_laptops['Company'] == 'Apple') ]
```

Encontraremos ahora las que sean Dell o Apple

```python
df_laptops[(df_laptops['Company'] == 'Dell') | (df_laptops['Company'] == 'Apple') ]
```

Encontraremos ahora las que sean Dell o Apple que cuesten más de 2000

```python
df_laptops[((df_laptops['Company'] == 'Dell') | (df_laptops['Company'] == 'Apple')) & (df_laptops['Price_euros']>2000) ]
```

## 4 Crear columna a travez de condiciones multiples

Usaremos el método de numpy llamado select

select(condiciones, valores que se tienen cuando se cumplen las condiciones)

```python
# Creamos condiciones
condiciones = [
    df_laptops['Price_euros']>3000,
    (df_laptops['Price_euros']<=3000) & (df_laptops['Price_euros']>2000),
    (df_laptops['Price_euros']<=2000) & (df_laptops['Price_euros']>800),
    df_laptops['Price_euros']<=800
]
#Creamos las etiquetas de nuestros rangos de valor
valores = ['Muy Caro','Caro', 'Barato', 'Muy Barato']
```

```python
df_laptops['Niveles de precio'] = np.select(condiciones, valores)
df_laptops.head(25)
```

```python
df_laptops.value_counts('Niveles de precio')
```

```python
condiciones2 = [
    df_laptops['Inches']>16,
    (df_laptops['Inches']<=16) & (df_laptops['Inches']>14),
    (df_laptops['Inches']<=14) & (df_laptops['Inches']>12),
    df_laptops['Inches']<=12
]

valores2 = ['Muy Grande','Grande', 'Pequeño', 'Muy Pequeño']

df_laptops['Screen size class'] = np.select(condiciones2, valores2)
```

```python
df_laptops.head(10)
```

```python
df_laptops.value_counts('Screen size class')
```

## 5 Método isin() : Filtrado simple


```python
#Obtener los que son HP o Apple
df_laptops[df_laptops['Company'].isin(['Apple','HP'])]
```

```python
df_laptops[df_laptops['Company'].isin(['Apple','HP'])].value_counts('Company')
```

## 6 Método isin() : Filtrado Múltiple

```python
# encontrar notebooks o ultrabooks de HP o Apple
filtro1 = df_laptops['TypeName'].isin(['Notebook','Ultrabook'])
filtro2 = df_laptops['Company'].isin(['Apple','HP'])
df_laptops[filtro1 & filtro2]
```

2do Ejemplo

Encontrar las laptops más caras y baratas, usando los métodos sort_values() y duplicated()

```python
# Ordenando por Company y Price de manera que lo más barato primero y lo más caro lo último
df_laptops=df_laptops.sort_values(['Company','Price_euros'])
```

```python
df_laptops
```

```python
df_laptops.value_counts('Company')  #todas las compañias tienen valores duplicados
```

```python
# Obtenemos el df con valores duplicados en la columna Company
df_laptops[df_laptops.duplicated('Company', keep= 'first')]  #este es el valor de keep que tiene por defecto
```

```python
# Obtenemos el df con valores NO duplicados en la columna Company
df_laptops[~df_laptops.duplicated('Company', keep= 'first')]
```

```python
# para solo ver las columnas Company y Price_euros
df_laptops[~df_laptops.duplicated('Company', keep= 'first')][['Company','Price_euros']]
#se mostraran solo los primeros valores que encuentre (por lo de keep first) e ignora el resto de los repetidos 
# y como esta en orden, siempre encuentra primero las de menor precio
```

```python
df_laptops[~df_laptops.duplicated('Company', keep= 'first')].value_counts('Company')
```

```python
# Lo mismo que antes, pero ahora con keep = 'last' para obtener las de mayor precio
df_laptops[~df_laptops.duplicated('Company', keep= 'last')][['Company','Price_euros']]
```

```python
# Lo mismo que antes pero ahora con keep = False, solo te regresa los que no tiene duplicados
df_laptops[~df_laptops.duplicated('Company', keep= False)]
# es como un dropduplicates, como en esa columna todos tienen repetidos no regresa nada
```

## 8 Eliminar duplicados con drop_duplicates()


```python
# Eliminar duplicados de la columna Company
df_laptops.drop_duplicates(['Company']).value_counts('Company') #contamos cuantos duplicados aparecen
```

```python
df_laptops.drop_duplicates(['Company'])[['Company', 'Price_euros']]
```

```python
# Ordenamos por compañia y precio
df_laptops=df_laptops.sort_values(['Company', 'Price_euros'])
```

```python
# A igual que antes, drop_duplicates, también tiene los mismos parametros de keep
df_laptops.drop_duplicates(['Company'], keep='first')[['Company','Price_euros']]
#este solo mantiene el primer duplicado
```

```python
df_laptops.drop_duplicates(['Company'], keep='last', ignore_index=True)[['Company','Price_euros']]
#ignore_index, esto reinicia el indice para esta visualización
```

```python
df_laptops.drop_duplicates(['Company'], keep=False)[['Company','Price_euros']]
```
