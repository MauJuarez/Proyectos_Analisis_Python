## 2.- Leer CSV de multiples URL con Pandas


- https://www.football-data.co.uk/mmz4281/2122/E0.csv
- https://www.football-data.co.uk/mmz4281/2122/E1.csv
- https://www.football-data.co.uk/mmz4281/2122/E2.csv
- https://www.football-data.co.uk/mmz4281/2122/E3.csv
- https://www.football-data.co.uk/mmz4281/2122/EC.csv

Por la estructura de los links, podemos crear un link dinámico.


Creamos una variable root o raíz

```python
root = "https://www.football-data.co.uk/mmz4281/"
```

Creamos una lista de links

```python
ligas = ["E0","E1","E2","E3","EC"]
```

Con un ciclo for, iremos creando los links de los demás 

```python
frames = []  #esta lista contendra los dataframes que vayamos creando

for liga in ligas:
  df= pd.read_csv(root+'2122/'+liga+'.csv', encoding ='unicode_escape')
  frames.append(df)
```

```python
len(frames)  # debe de haber 5 elementos
```

```python
frames[0]
```

Haremos lo mismo, pero para multiples temporadas, la idea es la misma.

```python
ligas = ["E0","E2","E3"]  # por facilidad, solo tomaremos estos 3
frames = []  #esta lista contendra los dataframes que vayamos creando

#aqui ira desde la temporada de 2015 hasta la de 2020
for liga in ligas:
  for season in range(15, 21):  # con esto adicionaremo el término para cada temporada
    df= pd.read_csv(root+str(season)+str(season+1) +'/'+liga+'.csv')
    df.insert(1,'season',season)  #se inserta esta columna en cada dataframe, para identificarlo
    frames.append(df)
```

Con el uso de insert, podemos especificar en que posición del dataframe queremos que se inserte la nueva columna

insert(posición, titulo de la columna, contenido de la columna)

```python
len(frames)  #debe de haber 18 elementos
```

```python
frames[0]
```

```python
frames[17]
```

## 3.- Organizando en un diccionario

Creamos el diccionario con las claves que usaremos

```python
dict_countries = { 'Spanish La Liga':'SP1', 'Spanish Segunda Division':'SP2',
              'German Bundesliga':'D1', 
              'English Premier League':'E0', 'English League 1':'E2', 'English League 2':'E3',
             }
```

```python
dict_countries['Spanish La Liga']
```

Haremos un ciclo sobre las llaves del diccionario.

Recordemos los ciclos sobre diccionarios

```python
#Este primero nos da las llaves
for league in dict_countries:
  print(league)
```

```python
#Este segundo nos da el contenido
for league in dict_countries:
  print(dict_countries[league])
```

Usando este diccionario (dict_countries), obtendremos para diferentes paises y no solo uno como hicimos antes.


```python
dict_historical_data ={}  # en este diccionario almacenaremos las llaves y los dataframes

# usando este diccionario, obtendremos para diferentes paises y no solo uno, como hicimos antes
for liga in dict_countries:
    frames = []  # se psiciona aquí para que cada ves que cambiemos de 'liga' esta se encuentre vacía 
    for season in range(15, 21): 
      df= pd.read_csv(root+str(season)+str(season+1) +'/'+dict_countries[liga]+'.csv')
      df.insert(1,'season',season)  
      frames.append(df)
    df_concat = pd.concat(frames)  #concatenamos los elementos de la lista frames, osea por cada "liga" uniremos las 6 temporadas
    dict_historical_data[liga] =  df_concat
```

```python
dict_historical_data.keys()  # las llaves
```

```python
dict_historical_data['Spanish La Liga']
```

Notamos que los dataframes de cada temporada fueron unidos, pues vemos que en la columna season van de 15 hasta la 20
