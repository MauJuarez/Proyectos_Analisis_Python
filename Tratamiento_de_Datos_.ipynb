Se eliminaran de nuestro modelo

```python
datos.dropna(subset=['Valor'], inplace=True)  # sin el inplace, no se modifica
```

De los visto en el boxplot, podemos ver que los outliers tienen valor mayor a 500,000 

```python
datos[datos['Valor']>=500000]
```

```python
valor = datos['Valor']
```

```python
Q1= valor.quantile(0.25)
Q3 = valor.quantile(0.75)
IIQ = Q3-Q1
lim_inf = Q1-1.5*IIQ
lim_sup = Q3+1.5*IIQ
```

```python
seleccion = ((valor >= lim_inf) & (valor <= lim_sup))  #con esta seleccion dejaremos fuera los outliers
```

```python
datos_new = datos[seleccion]   #crearemos un nuevo df con solo los valores correctos
```

```python
datos_new.boxplot(['Valor'])
```

Vamos a construir los limites pero por cierta columna

```python
datos.boxplot(['Valor'], by='Tipo') #esto hara el boxplot, pero por tipo
```

```python
grupo_tipo = datos.groupby('Tipo')
type(grupo_tipo)
```

```python
grupo_tipo = datos.groupby('Tipo')['Valor']
type(grupo_tipo)  # ya solo seria una serie, no un df, porque solo guardo la columna valor
```

```python
grupo_tipo.groups
```

```python
# crearemos las medidas del boxplot, pero para este grupo
Q1= grupo_tipo.quantile(0.25)
Q3 = grupo_tipo.quantile(0.75)
IIQ = Q3-Q1
lim_inf = Q1-1.5*IIQ
lim_sup = Q3+1.5*IIQ
```

```python
Q1 #PODEMOS VER QUE HIZO LA OPERACION PARA CADA TIPO 
```

```python
lim_inf
```

```python
lim_sup
```

```python
#creamos un nuevo df
datos_new = pd.DataFrame()
for tipo in grupo_tipo.groups.keys():
  etipo = datos['Tipo']  == tipo
  dentro_limite = (datos['Valor']>=lim_inf[tipo]) &(datos['Valor']<=lim_sup[tipo])
  seleccion = etipo & dentro_limite
  datos_seleccion = datos[seleccion]
  datos_new = pd.concat([datos_new, datos_seleccion])
```

Aclaraciones

Agruparemos los datos por e numero de cuartos como:

(1 y 2), (3 y 4 ), (5 y 6 ), (7 o mas)

```python
clases =[0,2,4,6,100] #aqui ponemos los limites de los cortes
cuartos = pd.cut(dataset['Cuartos'],clases)
```

```python
cuartos
```

En cada fila, le puso su clasificacion, que es lo que tiene en parentesis al lado, los NaN son los que tienen valor cero, porque para entrar en la categoria, debe ser superior al limite inferior

```python
pd.value_counts(cuartos)  #cuenta y trae el recuento por cada categoria
```

A continuacion creamos las etiquteas, para no tener los parentesis, pero deben ir en el orden que se pusieron los limites del cut

```python
milabel =['entre 1 y 2 cuartos','entre 3 y 4 cuartos','entre 5 y 6 cuartos','7 o mas']
cuartos = pd.cut(dataset['Cuartos'],clases,labels=milabel)
cuartos
```

```python
pd.value_counts(cuartos)
```

Notamos que hay demasiados registros con cuartos =0, entonces modifacaremos el cut

```python
cuartos = pd.cut(dataset['Cuartos'],clases,labels=milabel, include_lowest = True)
pd.value_counts(cuartos)   #incluyo el cero en la primera categoria
```
